## Random Walk

```python
import requests
from bs4 import BeautifulSoup
import datetime
import random
import re

random.seed(datetime.datetime.now().strftime('%s'))

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

def getLinks(articleUrl):
    try:
        response = requests.get(f'http://en.wikipedia.org{articleUrl}', headers=headers)
        response.raise_for_status() # Raise an exception for HTTP errors (4xx or 5xx)
        html = response.text
        bs = BeautifulSoup(html, 'html.parser')
        return bs.find('div', {'id':'bodyContent'}).find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))
    except requests.exceptions.HTTPError as err:
        print(f"HTTP error occurred: {err}")
        return [] # Return an empty list to prevent further errors
    except requests.exceptions.RequestException as err:
        print(f"An error occurred: {err}")
        return []

links = getLinks('/wiki/Kevin_Bacon')
while len(links) > 0:
    newArticle = links[random.randint(0, len(links)-1)].attrs['href']
    print(newArticle)
    links = getLinks(newArticle)
```

## Recursively crawling an entire site

```python
import requests
from bs4 import BeautifulSoup
import re
import time # Import time for delays

pages = set()

# Global counter and limit for pages visited
pages_visited_count = 0
MAX_PAGES_TO_VISIT = 50 # Set a reasonable limit to prevent infinite loops and 'Too Many Requests'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

def getLinks(pageUrl):
    global pages_visited_count

    # If the overall limit has been reached by any call in the recursion stack, return immediately.
    if pages_visited_count >= MAX_PAGES_TO_VISIT:
        return

    try:
        response = requests.get(f'http://en.wikipedia.org{pageUrl}', headers=headers)
        response.raise_for_status() # Raise an exception for HTTP errors (4xx or 5xx)
        html = response.text
        bs = BeautifulSoup(html, 'html.parser')

        for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
            if 'href' in link.attrs:
                newPage = link.attrs['href']
                if newPage not in pages:
                    # Crucial check: if adding this new page would exceed the limit, stop.
                    if pages_visited_count + 1 > MAX_PAGES_TO_VISIT:
                        print(f"Reached maximum pages to visit ({MAX_PAGES_TO_VISIT}). Stopping new page processing.")
                        return # Immediately return from the current call to stop processing its links

                    pages.add(newPage) # Mark as visited first
                    pages_visited_count += 1 # Then increment counter
                    print(f"Visiting page ({pages_visited_count}/{MAX_PAGES_TO_VISIT}): {newPage}")
                    time.sleep(0.5)

                    getLinks(newPage) # Recursive call

                    # After recursive call returns, check if the limit was reached by deeper calls.
                    # If so, propagate the return up the call stack.
                    if pages_visited_count >= MAX_PAGES_TO_VISIT:
                        return # Stop the current getLinks function from iterating further

    except requests.exceptions.HTTPError as err:
        print(f"HTTP error occurred for {pageUrl}: {err}")
    except requests.exceptions.RequestException as err:
        print(f"An error occurred for {pageUrl}: {err}")
    except Exception as err:
        print(f"An unexpected error occurred for {pageUrl}: {err}")

# Start crawling from the root of Wikipedia
print(f"Starting recursive link extraction, limited to {MAX_PAGES_TO_VISIT} pages.")
getLinks('/wiki/Main_Page') # Start with a common page like Main_Page or an empty string as before
print("Finished crawling.")
```

## Collecting Data Across an Entire Site

```python
import requests
from bs4 import BeautifulSoup
import re
import time # Import time for delays

pages = set()

# Global counter and limit for pages visited
pages_visited_count = 0
MAX_PAGES_TO_VISIT = 50 # Set a reasonable limit to prevent infinite loops and 'Too Many Requests'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

def getLinks(pageUrl):
    global pages_visited_count

    # If the overall limit has been reached by any call in the recursion stack, return immediately.
    if pages_visited_count >= MAX_PAGES_TO_VISIT:
        return

    try:
        response = requests.get(f'http://en.wikipedia.org{pageUrl}', headers=headers)
        response.raise_for_status() # Raise an exception for HTTP errors (4xx or 5xx)
        html = response.text
        bs = BeautifulSoup(html, 'html.parser')

        print(bs.h1.get_text())
        #mw-parser-output
        bodyContent = bs.find('div', {'id':'bodyContent'}).find_all('p')
        if len(bodyContent):
            print(bodyContent[0])
        # Some pages might not have 'ca-edit', so add a check
        edit_link_tag = bs.find(id='ca-edit')
        if edit_link_tag and edit_link_tag.find('a'):
            print(edit_link_tag.find('a').attrs['href'])

        for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
            if 'href' in link.attrs:
                newPage = link.attrs['href']
                if newPage not in pages:
                    # Crucial check: if adding this new page would exceed the limit, stop.
                    if pages_visited_count + 1 > MAX_PAGES_TO_VISIT:
                        print(f"Reached maximum pages to visit ({MAX_PAGES_TO_VISIT}). Stopping new page processing.")
                        return # Immediately return from the current call to stop processing its links

                    pages.add(newPage) # Mark as visited first
                    pages_visited_count += 1 # Then increment counter
                    print('-'*20)
                    print(f"Visiting page ({pages_visited_count}/{MAX_PAGES_TO_VISIT}): {newPage}")
                    time.sleep(0.5)

                    getLinks(newPage) # Recursive call

                    # After recursive call returns, check if the limit was reached by deeper calls.
                    # If so, propagate the return up the call stack.
                    if pages_visited_count >= MAX_PAGES_TO_VISIT:
                        return # Stop the current getLinks function from iterating further

    except requests.exceptions.HTTPError as err:
        print(f"HTTP error occurred for {pageUrl}: {err}")
    except requests.exceptions.RequestException as err:
        print(f"An error occurred for {pageUrl}: {err}")
    except AttributeError:
        print(f'This page ({pageUrl}) is missing something! Continuing.')
    except Exception as err:
        print(f"An unexpected error occurred for {pageUrl}: {err}")

print(f"Starting recursive link extraction, limited to {MAX_PAGES_TO_VISIT} pages.")
getLinks('/wiki/General-purpose_programming_language')
print("Finished crawling.")
```

## Crawling across the Internet

```python
import requests # Using requests instead of urllib.request
from urllib.parse import urlparse, urljoin # urljoin is crucial for proper relative URL handling
from bs4 import BeautifulSoup
import re
import datetime
import random
import time # Import time for delays

# Define headers for requests to avoid 403 Forbidden errors
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Retrieves a list of all Internal links found on a page
def getInternalLinks(bs, base_url):
    netloc = urlparse(base_url).netloc
    internalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc == '': # Relative path
            # Use urljoin to correctly resolve relative URLs to absolute URLs
            internalLinks.add(urljoin(base_url, link.attrs['href']))
        elif parsed.netloc == netloc: # Same domain, absolute path
            internalLinks.add(link.attrs['href'])
    return list(internalLinks)

# Retrieves a list of all external links found on a page
def getExternalLinks(bs, base_url):
    netloc = urlparse(base_url).netloc
    externalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        # Check if the netloc is different and not empty (i.e., external)
        if parsed.netloc != '' and parsed.netloc != netloc:
            externalLinks.add(link.attrs['href'])
    return list(externalLinks)

def getRandomExternalLink(startingPage):
    try:
        # Use requests.get instead of urlopen
        response = requests.get(startingPage, headers=headers, timeout=5)
        response.raise_for_status() # Raise an exception for HTTP errors (4xx or 5xx)
        bs = BeautifulSoup(response.text, 'html.parser')

        externalLinks = getExternalLinks(bs, startingPage)
        if not len(externalLinks):
            print('No external links, looking around the site for one')
            internalLinks = getInternalLinks(bs, startingPage)
            if internalLinks:
                # Add a delay to be polite to the server
                time.sleep(1)
                return getRandomExternalLink(random.choice(internalLinks))
            else:
                print(f"No internal links found on {startingPage} either. Stopping.")
                return None # No links to follow
        else:
            return random.choice(externalLinks)
    except requests.exceptions.HTTPError as err:
        print(f"HTTP error occurred for {startingPage}: {err}")
        return None
    except requests.exceptions.RequestException as err:
        print(f"An error occurred for {startingPage}: {err}")
        return None
    except Exception as err:
        print(f"An unexpected error occurred for {startingPage}: {err}")
        return None

def followExternalOnly(startingSite):
    link_count = 0
    max_links = 10 # Limit to prevent infinite loops and excessive requests
    current_site = startingSite

    while link_count < max_links:
        externalLink = getRandomExternalLink(current_site)
        if externalLink:
            print(f'Random external link is: {externalLink}')
            current_site = externalLink
            link_count += 1
            time.sleep(1) # Add a delay between requests
        else:
            print(f"No more external links to follow from {current_site}.")
            break
    print(f"Finished following {link_count} external links.")

# Removed direct call to avoid immediate execution in this context
followExternalOnly('https://www.oreilly.com/')
```

## Collect all External Links from a Site

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse, urljoin
import re
import time # Import time for delays

# Define headers for requests
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Recursion limit to prevent infinite loops for crawling
MAX_RECURSION_DEPTH = 50 # Set a reasonable limit
STOP_CRAWLING = False # Global flag to signal termination across all branches

# Retrieves a list of all Internal links found on a page
def getInternalLinks(bs, base_url):
    netloc = urlparse(base_url).netloc
    internalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc == '': # Relative path
            internalLinks.add(urljoin(base_url, link.attrs['href']))
        elif parsed.netloc == netloc: # Same domain, absolute path
            internalLinks.add(link.attrs['href'])
    return list(internalLinks)

# Retrieves a list of all external links found on a page
def getExternalLinks(bs, base_url):
    netloc = urlparse(base_url).netloc
    externalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        # Check if the netloc is different and not empty (i.e., external)
        if parsed.netloc != '' and parsed.netloc != netloc:
            externalLinks.add(link.attrs['href'])
    return list(externalLinks)


# Collects a list of all external URLs found on the site
allExtLinks = []
allIntLinks = []

def getAllExternalLinks(url, current_depth=0):
    global allExtLinks, allIntLinks, STOP_CRAWLING # Declare global variables

    if STOP_CRAWLING: # Check global stop flag
        return

    if current_depth > MAX_RECURSION_DEPTH:
        print(f"Max recursion depth ({MAX_RECURSION_DEPTH}) reached. Stopping for URL: {url}")
        STOP_CRAWLING = True # Set the flag to stop other branches
        return

    try:
        # Use requests.get instead of urlopen and include headers
        response = requests.get(url, headers=headers, timeout=5) # Added timeout for robustness
        response.raise_for_status() # Raise an exception for HTTP errors (4xx or 5xx)
        bs = BeautifulSoup(response.text, 'html.parser')

        internalLinks = getInternalLinks(bs, url)
        externalLinks = getExternalLinks(bs, url)

        for link in externalLinks:
            if STOP_CRAWLING: return # Check global flag before processing next link
            if link not in allExtLinks:
                allExtLinks.append(link)
                print(link)
        time.sleep(1) # Be polite, add a delay between requests
        if STOP_CRAWLING: return # Check global flag after delay

        for link in internalLinks:
            if STOP_CRAWLING: return # Check global flag before making a recursive call
            if link not in allIntLinks:
                allIntLinks.append(link)
                getAllExternalLinks(link, current_depth + 1) # Recursive call with incremented depth
                if STOP_CRAWLING: return # Check global flag after recursive call returns
    except requests.exceptions.HTTPError as err:
        print(f"HTTP error occurred for {url}: {err}")
    except requests.exceptions.RequestException as err:
        print(f"An error occurred for {url}: {err}")
    except Exception as err:
        print(f"An unexpected error occurred for {url}: {err}")

# Reset global variables before starting the crawl to ensure a fresh run
allExtLinks = []
allIntLinks = []
STOP_CRAWLING = False # Reset stop flag

allIntLinks.append('https://oreilly.com')
getAllExternalLinks('https://www.oreilly.com/', current_depth=0)
```
